```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import time

import os
os.environ["CUDA_VISIBLE_DEVICES"]='7'   # specify which GPU(s) to be used
```

# CC12M vs CC3M

## Processing CC3M

```python
path_dir = []
for dirname, _, filenames in os.walk('/home/myeongseob/clip-privacy/Open_clip_training/src/data/cc3m/'):
    for filename in filenames:
        #print(os.path.join(dirname, filename))
        if filename.split('.')[-1] == 'parquet':
            path_dir.append(os.path.join(dirname, filename))
```

```python
def collect_cc3M_data(path):
    df_cc3M = pd.read_parquet(path)
    N_cc3M = len(df_cc3M)  # number of samples from one parquet of laion 
    small_cc3M = df_cc3M.sample(n=N_cc3M)

    cc3M_images_array = small_cc3M["url"].iloc[0:N_cc3M].values
    return cc3M_images_array.tolist()
```

```python
## cal feature info 
start_time = time.time()

#processed_CC3M_text = []
processed_CC3M_image = []
for i in range(len(path_dir)):
    #cc3M_texts = collect_cc3M_data(path_dir[i])
    cc3M_images = collect_cc3M_data(path_dir[i])
    
    cc3M_images_re = []
    for val in cc3M_images:
        if val != None :
            cc3M_images_re.append(val)
    processed_CC3M_image.extend(cc3M_images_re)

print("--- %s seconds ---" % (time.time() - start_time))
```

```python
print(len(processed_CC3M_image))
processed_CC3M_image
```
